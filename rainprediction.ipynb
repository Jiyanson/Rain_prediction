```python
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler

# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Calculate silhouette scores for the scaled data
silhouette_vals = silhouette_samples(X_scaled, df_resampled['Cluster'])

# Calculate the average silhouette score
score = silhouette_score(X_scaled, df_resampled['Cluster'])

# Initialize the plot
y_ticks = []
y_lower, y_upper = 0, 0

plt.figure(figsize=(8, 6))

# Plot silhouette values for each cluster
for i, cluster in enumerate(np.unique(df_resampled['Cluster'])):
    cluster_silhouette = silhouette_vals[df_resampled['Cluster'] == cluster]
    cluster_silhouette.sort()
    y_upper += len(cluster_silhouette)
    plt.barh(range(y_lower, y_upper), cluster_silhouette, edgecolor='none', height=1)
    y_ticks.append((y_lower + y_upper) / 2)
    y_lower += len(cluster_silhouette)

# Add a vertical line at the average silhouette score
plt.axvline(x=score, color='red', linestyle='--')  # Reference line

# Adjust the ticks and labels
plt.yticks(y_ticks, np.unique(df_resampled['Cluster']))
plt.xlabel('Silhouette Coefficient')
plt.ylabel('Cluster')
plt.title('Silhouette Analysis for Weather Data Clustering')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X = df_resampled.drop('Rain_Tomorrow', axis=1)
y = df_resampled['Rain_Tomorrow']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# Step 3: Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Step 4: Train the linear regression model
linreg = LinearRegression()
linreg.fit(X_train_scaled, y_train)

# Step 5: Predict
y_pred = linreg.predict(X_test_scaled)

# Step 6: Evaluation
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error (MSE):", mse)
print("R-squared Score (R²):", r2)

# Optional: Visualizing actual vs. predicted values
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.6)
plt.xlabel("Actual Rain_Tomorrow")
plt.ylabel("Predicted Rain_Tomorrow")
plt.title("Linear Regression: Actual vs. Predicted")
plt.plot([0, 1], [0, 1], 'r--')  # Reference line if predicting perfectly
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import time

# 1. Définir le modèle
log_reg = LogisticRegression(max_iter=1000, solver='liblinear')

# 2. Grille des hyperparamètres
param_grid_logreg = {
    'C': [0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2']
}

# 3. Initialiser GridSearchCV
print("Initialisation de la recherche GridSearchCV pour Logistic Regression...")
start_time = time.time()
grid_logreg = GridSearchCV(estimator=log_reg, param_grid=param_grid_logreg,
                           cv=StratifiedKFold(n_splits=5), scoring='accuracy')

# 4. Entraînement
print("Début de l'entraînement du modèle Logistic Regression...")
grid_logreg.fit(X_train, y_train)
print(f"Entraînement terminé en {time.time() - start_time:.2f} secondes")

# 5. Meilleurs hyperparamètres
print("Meilleurs paramètres pour Logistic Regression : ", grid_logreg.best_params_)
print("Meilleure précision (CV) : ", grid_logreg.best_score_)

# 6. Prédictions
y_pred_logreg = grid_logreg.predict(X_test)

# 7. Évaluation
print("Logistic Regression Accuracy: ", accuracy_score(y_test, y_pred_logreg))
print(classification_report(y_test, y_pred_logreg))

# 8. Matrice de confusion
conf_matrix_logreg = confusion_matrix(y_test, y_pred_logreg)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_logreg, annot=True, fmt='d', cmap='Blues',
            xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
plt.title("Matrice de confusion - Logistic Regression")
plt.xlabel("Valeurs prédites")
plt.ylabel("Valeurs réelles")
plt.show()

# 9. Visualisation des scores
scores_logreg = grid_logreg.cv_results_['mean_test_score']
param_combinations_logreg = [
    f"C={p['C']}, penalty={p['penalty']}" for p in grid_logreg.cv_results_['params']
]

plt.figure(figsize=(12, 8))
plt.barh(param_combinations_logreg, scores_logreg, color='skyblue')
plt.title("Précision moyenne - Logistic Regression (Grid Search)")
plt.xlabel("Précision moyenne")
plt.ylabel("Hyperparamètres")
plt.grid(axis='x')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# Initialisation du modèle Naïve Bayes
nb = GaussianNB()

# Grille de recherche des hyperparamètres
param_grid_nb = {
    'var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6, 1e-5]  # Plage de valeurs pour var_smoothing
}

# GridSearchCV avec validation croisée stratifiée
grid_nb = GridSearchCV(
    estimator=nb,
    param_grid=param_grid_nb,
    cv=StratifiedKFold(n_splits=5),  # Validation croisée avec 5 groupes
    scoring='accuracy'  # Critère d'évaluation : précision
)

# Entraîner le modèle avec la grille de recherche
grid_nb.fit(X_train, y_train)

# Meilleurs hyperparamètres et meilleure précision
print("Meilleurs paramètres pour Naïve Bayes : ", grid_nb.best_params_)
print("Meilleure précision : ", grid_nb.best_score_)

# Prédictions sur les données de test avec les meilleurs hyperparamètres
y_pred_nb = grid_nb.predict(X_test)

# Évaluation des résultats
print("Naïve Bayes Accuracy: ", accuracy_score(y_test, y_pred_nb))
print(classification_report(y_test, y_pred_nb))

# Ajout d'affichages graphiques

# 1. Matrice de confusion
conf_matrix = confusion_matrix(y_test, y_pred_nb)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
plt.title("Matrice de confusion - Naïve Bayes")
plt.xlabel("Valeurs prédites")
plt.ylabel("Valeurs réelles")
plt.show()

# 2. Évolution des scores avec les hyperparamètres
scores = grid_nb.cv_results_['mean_test_score']
param_values = param_grid_nb['var_smoothing']

plt.figure(figsize=(8, 6))
plt.plot(param_values, scores, marker='o', linestyle='-', color='b')
plt.xscale('log')  # Échelle logarithmique pour var_smoothing
plt.title("Évolution de la précision avec var_smoothing")
plt.xlabel("var_smoothing (échelle log)")
plt.ylabel("Précision moyenne")
plt.grid(True)
plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Step 1: Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Step 2: Define the SVM model
svm = SVC()

# Step 3: Grid search for hyperparameter tuning
param_grid = {
    'C': [1, 10],  # Reduced for speed
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale']
}
cv = StratifiedKFold(n_splits=5)
grid_search = GridSearchCV(svm, param_grid, cv=cv, scoring='accuracy', n_jobs=-1, verbose=0)
grid_search.fit(X_train_scaled, y_train)

# Step 4: Best model and evaluation
best_svm = grid_search.best_estimator_
y_pred = best_svm.predict(X_test_scaled)

print("Best Parameters:", grid_search.best_params_)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# Step 5: Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Purples')
plt.title("Confusion Matrix - SVM")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import time
import numpy as np

# Step 1: Define the model
xgb = XGBClassifier(eval_metric='logloss', random_state=42)

# Step 2: Define the hyperparameter grid
param_grid_xgb = {
    'n_estimators': [50, 100],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0]
}

# Step 3: Set up GridSearchCV
grid_xgb = GridSearchCV(
    estimator=xgb,
    param_grid=param_grid_xgb,
    cv=StratifiedKFold(n_splits=5),
    scoring='accuracy',
    n_jobs=-1
)

# Step 4: Train the model
print("Training XGBoost model with grid search...")
start_time = time.time()
grid_xgb.fit(X_train, y_train)
print(f"Training completed in {time.time() - start_time:.2f} seconds")

# Step 5: Evaluation
print("Best hyperparameters: ", grid_xgb.best_params_)
print("Best cross-validated accuracy: ", grid_xgb.best_score_)

# Predict on test data
y_pred_xgb = grid_xgb.predict(X_test)

print("Test Accuracy: ", accuracy_score(y_test, y_pred_xgb))
print(classification_report(y_test, y_pred_xgb))

# Confusion matrix
conf_matrix_xgb = confusion_matrix(y_test, y_pred_xgb)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_xgb, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
plt.title("Confusion Matrix - XGBoost")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Optional: Visualize best model’s feature importance
importances = grid_xgb.best_estimator_.feature_importances_
features = X_train.columns

plt.figure(figsize=(10, 6))
sns.barplot(x=importances, y=features)
plt.title('Feature Importance - XGBoost')
plt.xlabel('Importance')
plt.ylabel('Features')
plt.tight_layout()
plt.show()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import time

# Définir le modèle
mlp = MLPClassifier(max_iter=300, random_state=42)

# Grille réduite pour RandomizedSearchCV
param_dist_mlp = {
    'hidden_layer_sizes': [(50, 50), (100,), (100, 50)],
    'learning_rate': ['constant', 'adaptive'],
    'alpha': [0.0001, 0.001],
    'solver': ['adam', 'sgd']
}

# RandomizedSearchCV (10 combinaisons aléatoires max)
print("Initialisation de RandomizedSearchCV...")
start_time = time.time()
random_mlp = RandomizedSearchCV(
    estimator=mlp,
    param_distributions=param_dist_mlp,
    n_iter=10,  # Try only 10 random combinations
    cv=3,
    scoring='accuracy',
    random_state=42,
    n_jobs=-1
)

# Entraîner le modèle
print("Début de l'entraînement du modèle (MLP)...")
random_mlp.fit(X_train, y_train)
print(f"Entraînement terminé en {time.time() - start_time:.2f} secondes")

# Évaluation
print("Meilleurs paramètres : ", random_mlp.best_params_)
print("Meilleure précision moyenne : ", random_mlp.best_score_)

y_pred_mlp = random_mlp.predict(X_test)

print("Accuracy test : ", accuracy_score(y_test, y_pred_mlp))
print(classification_report(y_test, y_pred_mlp))

# Matrice de confusion
conf_matrix_mlp = confusion_matrix(y_test, y_pred_mlp)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_mlp, annot=True, fmt='d', cmap='Oranges', xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
plt.title("Matrice de confusion - MLP (optimisée)")
plt.xlabel("Prédit")
plt.ylabel("Réel")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import time

# Step 1: Define the model
rf = RandomForestClassifier(random_state=42)

# Step 2: Define the hyperparameter grid
param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'bootstrap': [True, False]
}

# Step 3: Grid Search with Stratified K-Fold CV
grid_rf = GridSearchCV(
    estimator=rf,
    param_grid=param_grid_rf,
    cv=StratifiedKFold(n_splits=5),
    scoring='accuracy',
    n_jobs=-1
)

# Step 4: Train the model
print("Training Random Forest model with grid search...")
start_time = time.time()
grid_rf.fit(X_train, y_train)
print(f"Training completed in {time.time() - start_time:.2f} seconds")

# Step 5: Evaluation
print("Best hyperparameters: ", grid_rf.best_params_)
print("Best cross-validated accuracy: ", grid_rf.best_score_)

# Predict on test data
y_pred_rf = grid_rf.predict(X_test)

print("Test Accuracy: ", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))

# Step 6: Confusion matrix
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_rf, annot=True, fmt='d', cmap='Greens', xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
plt.title("Matrice de confusion - Random Forest")
plt.xlabel("Valeurs prédites")
plt.ylabel("Valeurs réelles")
plt.show()

# Step 7: Feature importance
importances = grid_rf.best_estimator_.feature_importances_
features = X_train.columns

plt.figure(figsize=(10, 6))
sns.barplot(x=importances, y=features)
plt.title('Importance des variables selon Random Forest')
plt.xlabel('Importance')
plt.ylabel('Variables')
plt.tight_layout()
plt.show()
```
